## 남자 의사 사망자 수 모듈화

```python
def death_man(array):
    death_man = pd.read_excel('../../data/사망확률남자.xlsx')
    death_man = death_man[7::]
    death_man = death_man.set_index(['연도'])
    data_man_death = death_man*array
    return data_man_death
data_man_death = death_man(Grad_man)
data_man_death = data_man_death.iloc[:,25:80]
data_man_death = data_man_death.round()
data_man_death
```

## 생존자 의사수

```python
Grad_man = Grad_man.iloc[:,25:80]
servied_man = Grad_man - data_man_death
```
