```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import plotly.graph_objects as go
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
import plotly.offline as pyo
```

```python
manDoc = pd.read_csv('../../data/totalManNumber.csv')
index = list(range(1955,2048))
manDoc = manDoc.set_index([index])
manDoc
```

```python
womDoc = pd.read_csv('../../data/totalWomNumber.csv')
index = list(range(1955,2048))
womDoc = womDoc.set_index([index])
womDoc
```

```python
def slicingPerson(year):
    value = []
    value.append(manDoc.iloc[year])
    value.append(womDoc.iloc[year])
    return value

def slicingPersonAndSum(year):
    value = []
    value.append(manDoc.iloc[year].sum())    
    value.append(womDoc.iloc[year].sum())
    return value

def makeDataFrame():
    manDocNum  = pd.DataFrame(columns=['index','man', 'wom', 'both'])
    for i in range(1950, 2048):
        yearIndex = i-1955
        man = slicingPersonAndSum(yearIndex)[0]
        wom = slicingPersonAndSum(yearIndex)[1]
        both = man+wom
        appendNext = pd.DataFrame(data = [[i, man, wom, both]], columns= ['index','man', 'wom', 'both'])
        manDocNum = manDocNum.append(appendNext)
        i += 1
    manDocNum.set_index('index')    
    return manDocNum
    
dataFrame = makeDataFrame()
manDocNum = list(dataFrame['man'])
manDocNum
```

```python
#1985년은 slicing기준으로 30.

def makeSumFigure(year):
    yearValue = int(year)
    yearIndex = yearValue-1955

    dataFrmae = makeDataFrame()
    year = dataFrame.index.tolist()
    manDocNum = list(dataFrame['man'])
    womDocNum =list(dataFrame['wom'])
    bothDocYear = list(dataFrame['both'])
    
    
    trace1 = go.Bar(name='Man', x=manDocNum, y=year)
    trace2 = go.Bar(name = 'Woman', x=womDocNum, y=year)
    trace3 = go.Bar(name = 'Both', x=bothDocYear, y = year)
    
    data = [trace1, trace2, trace4]
    layout = go.Layout(title=year+'년 연령별 의사 수')
    fig = go.Figure(data=data, layout=layout)

    fig.update_layout(
        height=500,
        width=1000,
        paper_bgcolor='rgb(248, 248, 255)',
        plot_bgcolor='rgb(248, 248, 255)',
        margin=dict(l=0),
        showlegend=False,
    )
    return fig

makeSumFigure('2018').show()
```
