```python
# 분류모델함수 생성
# x - x값, 네 - 시작점, xl - x축 조정, yl - y축 조정
def sigmoid_function(x, sp, xl, yl):
    
    result = 1 / ( yl+np.exp((x-(sp+(10/xl/2)))*xl) )
    
    for i in range(np.size(x)):
        if x[i]<sp:
            result[i]=0
        if x[i]>(sp+(10/xl)):
           result[i]=0
            
    return result
                                     
```

```python
# 분류모델 확률질량함수 생성

def sigmoid_model(sp, xl, yl):
    x = np.arange(1, 100, 1)
    y = sigmoid_function(x, sp, xl, yl)
    sum = np.sum(y)
    z = y/sum
    
    return z
```

```python
# 분배모델 시각화
x = np.arange(0, 100, 0.1)
y = sigmoid_function(x, 20, 0.5, 0.01)
plt.plot(x, y)
plt.ylim(-0.1, 100)
plt.xlim(0, 50)
plt.show()
```

```python
model1 = sigmoid_model(26,0.5,0.01) # 의대 남
model2 = sigmoid_model(26,0.5,0.01) # 의대 여
model3 = sigmoid_model(28,0.5,0.01) # 의전원 남
model4 = sigmoid_model(28,0.5,0.01) # 의전원 여
model5 = sigmoid_model(27,0.5,0.01) # 재시험 남
model6 = sigmoid_model(27,0.5,0.01) # 재시험 여

data1 = np.around(model1*npData[0,0])

print(str(npData[0,0]) + ' ' + str(np.sum(data1)))
print(data1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 0
variance = 5
sigma = math.sqrt(variance)
x = np.linspace(mu - 2*sigma, mu + 2*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 0, 0.1 # mean and standard deviation

# np.random.nomral 함수를 이용해서 평균 0, 표준편차 0.1인 sample들을 1000개 추출한다.
s = np.random.normal(mu, sigma, 1000)

#s = npData2


# sample들의 historgram을 출력한다.
count, bins, ignored = plt.hist(s, 50, density = True)


# sample들을 이용해서 Gaussian Distribution의 shape을 재구축해서 line으로 그린다.
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.show()
```

```python
import scipy.stats as st
import numpy as np
a = 2; s =0.5
y=st.gamma.rvs(a=a,scale=s, size=500) #Gamma(2,0.5) size=500
st.describe(y) #elelmentary statistics
n, rgn, mean, var, skew, kurt= st.describe(y)
print('min= %10.4f, mean= %10.4f, varince=%10.4f' % (rgn[0], mean,var))
mean, var, skew, kurt = st.gamma.stats(a=a,scale=s, moments='mvsk')
print(mean, var)
st.gamma.moment(n=2, a=a,scale=s) #2nd Moment
st.gamma.moment(n=1, a=a,scale=s) #1nd Moment
st.gamma.median(a=a,scale=s)
st.gamma.std(a=a,scale=s)
a, loc, scale=st.gamma.fit(y)
print('alpha(shape)= %6.3f, scale= %6.3f, ' % (a, scale))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gamma

fig, ax = plt.subplots(1, 1)
x = np.linspace(gamma.ppf(0.001,a=a,scale=s),gamma.ppf(0.999,a=a,scale=s), 100)
ax.set_title('Population PDF & Histigram of Generating data')

ax.plot(x, gamma.pdf(x,a=a,scale=s), 'r--', lw=2, alpha=0.6,label='Gamma pdf 1 ('+str(a)+', '+str(s)+')')
ax.plot(x, gamma.pdf(x,a=0.8*a,scale=s), 'b--', lw=2, alpha=0.6,label='Gamma pdf 2 ('+str(0.8*a)+', '+str(s)+')')
ax.plot(x, gamma.pdf(x,a=1.2*a,scale=s), 'k--', lw=2, alpha=0.6,label='Gamma pdf 3 ('+str(1.2*a)+', '+str(s)+')')

ax.hist(y, density=True, histtype='stepfilled', alpha=0.2)

ax.legend(loc='best', frameon=False)
plt.show()
```
