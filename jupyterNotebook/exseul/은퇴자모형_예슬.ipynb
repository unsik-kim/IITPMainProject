## 은퇴자 모형_감마분포

```python
#감마분포_연속형
from numpy import linspace, exp

x = linspace(1, 20, 1000)
x

from scipy.special import gamma

#shape와 scale에 따라 변경

shape = 30
scale = .3
y = x ** (shape - 1) * exp(-x / scale) / (scale ** shape * gamma(shape))
y

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('Gamma Distribution(shape = 10, scale = .5, loc = 0)')
plt.grid()
plt.show()
```

```python
def draw_fractal(func, max_step, startX = 0, startY = 0):
    x, y = startX, startY
    x_list = [x]
    y_list = [y]

    for i in range(max_step):
        x, y = func(x, y)
        x_list.append(x)
        y_list.append(y)
        
        plt.plot(x_list, y_list)
    plt.show()
    
import random
import matplotlib.pyplot as plt
def simple_two_point(x, y):
    for i in range(2):
        return x+1, y+1 if x == 3 else -y+1

draw_fractal(simple_two_point, 2)
```

## 은퇴자 모형_이항정규분포

```python
from scipy.special import factorial, comb

n = 20
x = 5

f = factorial(n) / (factorial(x) * factorial(n - x))
print('factorial()을 이용한 콤비네이션 연산 결과: ' + str(f))

c = comb(20, 5)
print('comb()을 이용한 콤비네이션 연산 결과: ' + str(c))

from numpy import arange

x = arange(21)
x

p = 1 / 3
y = comb(n, x) * p ** x * (1 - p) ** (n - x)
y

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12, 8))
plt.plot(x, y, 'o--')
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('Binomial Distribution(n = 20, p = 1/3)')
plt.grid()
plt.show()


#참고 : https://blog.naver.com/PostView.nhn?blogId=parksehoon1971&logNo=221584597915&parentCategoryNo=&categoryNo=&viewDate=&isShowPopularPosts=false&from=postView
```

# 정규분포 모양 은퇴율 모형

```python
#함수 정의 X
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math

#변경가능한 파라미터는  mu와 variance
mu = 66.1
variance = 30
def normaldist_model(mu, variance):
    sigma = math.sqrt(variance)
    x = np.linspace(0,100, 8500)
    plt.plot(x, 5*stats.norm.pdf(x, mu, sigma))
    plt.show()
```

```python
# 분류모델 확률질량함수 생성

def sigmoid_model(sp, xl, yl):
    x = np.arange(1, 101, 1)
    y = sigmoid_function(x, sp, xl, yl)
    sum = np.sum(y)
    z = y/sum
    
    return z
```

```python
# 분류모델함수 생성
# x - x값, 네 - 시작점, xl - x축 조정, yl - y축 조정
def sigmoid_function(x, sp, xl, yl):
    
    result = 1 / ( yl+np.exp((x-(sp+(10/xl/2)))*xl) )
    
    for i in range(np.size(x)):
        if x[i]<sp:
            result[i]=0
        if x[i]>(sp+(10/xl)):
           result[i]=0
            
    return result
```
