```python
# 분류모델함수 생성
# x - x값, 네 - 시작점, xl - x축 조정, yl - y축 조정
def sigmoid_function(x, sp, xl, yl):
    
    result = 1 / ( yl+np.exp((x-(sp+(10/xl/2)))*xl) )
    
    for i in range(np.size(x)):
        if x[i]<sp:
            result[i]=0
        if x[i]>(sp+(10/xl)):
           result[i]=0
            
    return result
                                     
```

```python
# 분류모델 확률질량함수 생성

def sigmoid_model(sp, xl, yl):
    x = np.arange(1, 100, 1)
    y = sigmoid_function(x, sp, xl, yl)
    sum = np.sum(y)
    z = y/sum
    
    return z
```

```python
# 분배모델 시각화
x = np.arange(0, 100, 0.1)
y = sigmoid_function(x, 20, 0.5, 0.01)
plt.plot(x, y)
plt.ylim(-0.1, 100)
plt.xlim(0, 50)
plt.show()
```

```python
model1 = sigmoid_model(26,0.5,0.01) # 의대 남
model2 = sigmoid_model(26,0.5,0.01) # 의대 여
model3 = sigmoid_model(28,0.5,0.01) # 의전원 남
model4 = sigmoid_model(28,0.5,0.01) # 의전원 여
model5 = sigmoid_model(27,0.5,0.01) # 재시험 남
model6 = sigmoid_model(27,0.5,0.01) # 재시험 여

data1 = np.around(model1*npData[0,0])

print(str(npData[0,0]) + ' ' + str(np.sum(data1)))
print(data1)
```
