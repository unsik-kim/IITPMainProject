```python
import pandas as pd
import numpy as np

pd.set_option('display.max_row', 1200)
pd.set_option('display.max_columns', 100)
```

```python
# 1950~2067년
examResult = (pd.read_excel('output.xlsx'))
```

```python
examResult
```

```python
examResult1 = examResult[examResult['연도']==2020]
```

```python
examResult2 = examResult1[examResult1['직종']!='의사']
```

```python
examResult3 = examResult2[examResult2['성별']=='남']
```

```python
examResult4 = examResult3[['연령대','값']].groupby(['연령대']).mean()
```

```python
np.array(examResult4).T[0]
```

```python
resultList = [[],[]]
for i in range(2001,2021):
    examResult1 = examResult[examResult['연도']==i]
    examResult2 = examResult1[examResult1['성별']=='남']
    examResult3 = examResult2[['연령대','값']].groupby(['연령대']).mean()
    resultList[0].append(np.array(examResult3).T[0])
for i in range(2001,2021):
    examResult4 = examResult[examResult['연도']==i]
    examResult5 = examResult4[examResult4['성별']=='여']
    examResult6 = examResult5[['연령대','값']].groupby(['연령대']).mean()
    resultList[1].append(np.array(examResult6).T[0])
    
dfManPassValue = pd.DataFrame(resultList[0])
dfWomanPassValue = pd.DataFrame(resultList[1])
dfManPassValue.columns = ['20','30','40','50','60']
dfWomanPassValue.columns = ['20','30','40','50']
dfManPassValue['sex'] = list([ '남' for i in range(20) ])
dfWomanPassValue['sex'] = list([ '여' for i in range(20) ])
dfManPassValue.index = range(2001,2021)
dfWomanPassValue.index = range(2001,2021)
dfPassValue = pd.concat([dfManPassValue,dfWomanPassValue]).fillna(0)
```

```python
dfPassValue.to_excel('output.xlsx')
```
