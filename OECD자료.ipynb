```python
import pandas as pd
import matplotlib.pyplot as plt

data_df = pd.read_excel('data/OECD의사수.xlsx')
# Variable 0:Practising physicians / 1:Professionally active physicians / 2:Physicians licensed to practice
# Measure 0:Number of persons (head counts) / 1: Density per 1 000 population (head counts)
# OECD 0: OECD Economies / 1: Non-OECD Economies
print(data_df)
```

```python
meanData_df = data_df.groupby(['Variable','Measure','OECD'],dropna=True).mean()
```

```python
meanData_df
```

```python
is_koreaData = data_df['Country']=='Korea'
koreaData = data_df[is_koreaData]
koreaData
```

```python
y1 = meanData_df.iloc[2]
y2 = meanData_df.iloc[3]
y3 = meanData_df.iloc[5]
y4 = meanData_df.iloc[7]
y5 = koreaData.iloc[1,4:]
y6 = koreaData.iloc[3,4:]
print(y2)
```

```python
yearList = range(2000,2020)
plt.plot(yearList, y1,label='Practising physicians/OECD', color='blue', alpha = 0.2)  # label 설정값을 legend에 나타날 수 있음
plt.plot(yearList, y2, label='Practising physicians/NONE_OECD', color='blue', alpha = 0.5)
plt.plot(yearList, y5, label='Practising physicians/Korea', color='blue', alpha = 1)
plt.plot(yearList, y4, label='Physicians licensed to practice', color='orange', alpha = 0.3)
plt.plot(yearList, y6, label='Physicians licensed to practice/Korea', color='orange', alpha = 1)

plt.xlabel("Year")
plt.ylabel("Density per 1 000 population")

plt.grid()
plt.legend()

plt.show()
```
