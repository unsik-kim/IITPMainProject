```python
doct = pd.read_excel('data/시도별 의료인력 현황.xlsx',thousands=',')
```

```python
orientalDoct = doct[doct["의료인력별"]=="한의사"]
orientalDoct
```

```python
orientalDoct.drop(['항목'], axis='columns', inplace=True)
orientalDoct.drop(['단위'],axis='columns', inplace=True)
orientalDoct= orientalDoct.fillna(0)
orientalDoct = orientalDoct.replace('-', 0)
```

```python
orientalDoct.columns[2][0:4]
```

```python
# 지역 데이터 업로드
test_list = []
for i in range(len(orientalDoct.index)):
    local = orientalDoct.iloc[i,0]
    human = orientalDoct.iloc[i,1]
    for j in range(len(orientalDoct.columns[2:])):
        year = orientalDoct.columns[j+2]
        value = orientalDoct.iloc[i,j+2]
        dic_doc = {"의료인력":human,"시도":local,"년도":int(year[0:4]),"값":int(value)}
        test_list.append(dic_doc)

print(type(test_list))
print(type(test_list[0]))
print(type(test_list[0]['년도']))
print(type(test_list[0]['시도']))
```
