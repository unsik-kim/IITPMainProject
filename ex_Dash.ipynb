```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib


matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
subject_doct = pd.read_excel('시도별_전문과목별_전문의_현황_2005-2018_2.xlsx',thousands = ',')

dataDict = {}

for i in range(2005,2019):
    data = subject_doct[subject_doct['년도']==i]
    data.fillna(0)
    data.reset_index(drop=True)
    data.drop(['년도'], axis='columns', inplace=True)
    data.set_index(data['과목'],inplace = True)
    data.drop(['과목'],axis='columns', inplace=True)                 
    dataDict.setdefault(str(i), data)
    dataDict[str(i)] = dataDict[str(i)].fillna(0)
    dataDict[str(i)] = dataDict[str(i)].replace('-', 0)
    dataDict[str(i)].drop(['계'], inplace=True)  
    
dataDict['2015']
```

```python
# 과목별 계를 기준으로 sorting
sorted_dataDict = {}

for i in range(2005,2019):
    sorted_dataDict.setdefault(str(i), dataDict[str(i)].sort_values(by='계' ,ascending=False))
```

```python
sorted_dataDict['2006']
```

```python
import plotly.express as px


fig = px.bar(sorted_dataDict['2017'],x=sorted_dataDict['2017'].columns[range(1,18)], y=sorted_dataDict['2017'].index, orientation='h',
             height=1000,width=1400,
             title='2018년 시도별 전공의 수')

fig.update_layout(barmode='stack',
                  margin=dict(l=0, r=0, t=100, b=0),
                  font=dict(family='Malgun Gothic', size=13,color='rgb(67, 67, 67)'),
                  legend_title_text='       도시',
                xaxis = dict(
                      title = '',
                   ),
                 yaxis = dict(
                      title = '',
                      titlefont = dict(
                         size = 1
                      )
                 )
            )

fig.show()
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px

import pandas as pd

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = dash.Dash()

app.layout = html.Div([
    dcc.Graph(id='graph-with-slider'),
    dcc.Slider(
        id='year-slider',
        min=2005,
        max=2018,
        value=2010,
        marks={i: '{}'.format(i) for i in range(2005,2019)},
        step=1
    )
])


@app.callback(
    Output('graph-with-slider', 'figure'),
    [Input('year-slider', 'value')])
def update_figure(selected_year):
    year = str(selected_year)
    fig = px.bar(sorted_dataDict[year],x=sorted_dataDict[year].columns[range(1,18)], y=sorted_dataDict[year].index, orientation='h',
                 height=900,
                 title=year+'년 시도별 전공의 수')

    fig.update_layout(barmode='stack',
                      margin=dict(l=0, r=0, t=100, b=0),
                      font=dict(family='Malgun Gothic', size=13,color='rgb(67, 67, 67)'),
                      legend_title_text='       도시',
                    xaxis = dict(
                          title = '',
                       ),
                     yaxis = dict(
                          title = '',
                          titlefont = dict(
                             size = 1
                          )
                     )
                )
    return fig

app.run_server(debug=True, use_reloader=False)  # Turn off reloader if inside Jupyter
```
