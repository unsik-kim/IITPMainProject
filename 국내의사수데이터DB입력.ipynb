```python
import pandas as pd
import matplotlib.pyplot as plt
import connectDB

myclient = connectDB.connectDB()

newDB = myclient['FriDB']
newCollection = newDB['MedicalPersonData']


data_df1 = pd.read_csv('data/국내의료기관종사인력/데이터1.csv', encoding='cp949')
data_df1
```

```python
newDF1 = data_df1.groupby(['행정구분','직종구분','데이터구분']).sum()


newDF1 = newDF1.reset_index()
newDF1
```

```python
resultList1 = []

for i in range(len(newDF1.index)):
    for j in range(len(newDF1.columns[3:])):
        local = newDF1.iloc[i,0]
        subject = newDF1.iloc[i,1]
        dataType = newDF1.iloc[i,2]
        value = newDF1.iloc[i,j+3]
        year = newDF1.columns[j+3]
        
        resultDict1 = {'년도':int(year[0:4]), '행정구분': local, '직종구분':subject, '데이터구분':dataType, '값':round(float(value),3)}
        
        resultList1.append(resultDict1)
        
len(resultList1)

resultList1
```

```python
newCollection.insert_many(resultList1)
```
