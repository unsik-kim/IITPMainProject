```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
import connectDB
import pandas as pd
import plotly.graph_objects as go

# DB 연결
myclient = connectDB.connectDB()

myDB = myclient['FriDB']
popCol = myDB['Population2']
csdCol = myDB['CitySubjectDoc']

popDF = pd.DataFrame(popCol.find({}, {'_id': 0}))
csdDF = pd.DataFrame(csdCol.find({}, {'_id': 0}))

popAllDF = popDF[popDF['행정기관'] == '전국']
popAllDF = popAllDF[popDF['구분'] == '총인구수']
resultDF = popDF.iloc[13:, 2:4]
pd.DataFrame.reset_index(resultDF).iloc[:,1:]
csdDF = pd.DataFrame(csdCol.find({}, {'_id': 0}))
csdDF

popLocDF = popDF[popDF['행정기관'] != '전국']
popLocDF = popLocDF[popDF['구분'] == '총인구수']
csdDF

csdDF.head(20)
```

```python
resultList = []
for i in range(len(csdDF.index)):
    subj = csdDF.iloc[i,0]
    year = csdDF.iloc[i,1]
    local = csdDF.iloc[i,2]
    value = csdDF.iloc[i,3]
    
    if local == '강원':
        local = '강원도'
    elif local == '경기':
        local = '경기도'
    elif local == '경남':
        local = '경상남도'
    elif local == '경북':
        local = '경상북도'
    elif local == '광주':
        local = '광주광역시'
    elif local == '대구':
        local = '대구광역시'
    elif local == '대전':
        local = '대전광역시'
    elif local == '부산':
        local = '부산광역시'
    elif local == '서울':
        local = '서울특별시'
    elif local == '세종':
        local = '세종특별자치시'
    elif local == '울산':
        local = '울산광역시'
    elif local == '인천':
        local = '인천광역시'
    elif local == '전남':
        local = '전라남도'
    elif local == '전북':
        local = '전라북도'
    elif local == '제주':
        local = '제주특별자치도'
    elif local == '충북':
        local = '충청북도'
    elif local == '충남':
        local = '충청남도'
    
    popDataDF = popLocDF[popLocDF['행정기관']==local]
    popDataDF = popDataDF[popDataDF['년도']==year]
    popDF = popDataDF.iloc[0,2]
    perDoc = value/popDF*1000
    perDoc = perDoc if perDoc!=None else 0
    #print('{} - {}, {}, {}, {}, {}, {}'.format(i, subj, year, local, value, popDataDF.iloc[0,2], perDoc))
    resultDict = {'과목':subj,'년도':int(year),'지역':local,'의사수':int(value),'인구수':int(popDF),'값':round(perDoc,3)}
    resultList.append(resultDict)
    
resultList

```
